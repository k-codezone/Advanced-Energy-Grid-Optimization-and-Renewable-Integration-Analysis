````python
# Forecasting with LSTM
def lstm_forecast(data, look_back=1):
    # Normalize data
    scaler = MinMaxScaler(feature_range=(0, 1))
    dataset = scaler.fit_transform(data.values.reshape(-1, 1))

    # Split into train and test sets
    train_size = int(len(dataset) * 0.8)
    train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]

    # Create dataset for LSTM
    def create_dataset(dataset, look_back=1):
        X, Y = [], []
        for i in range(len(dataset)-look_back-1):
            a = dataset[i:(i+look_back), 0]
            X.append(a)
            Y.append(dataset[i + look_back, 0])
        return np.array(X), np.array(Y)

    X_train, y_train = create_dataset(train, look_back)
    X_test, y_test = create_dataset(test, look_back)

    # Reshape input to be [samples, time steps, features]
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    # Create and fit the LSTM network
    model = Sequential()
    model.add(LSTM(4, input_shape=(look_back, 1)))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    model.fit(X_train, y_train, epochs=20, batch_size=1, verbose=2)

    # Make predictions
    train_predict = model.predict(X_train)
    test_predict = model.predict(X_test)

    # Invert predictions
    train_predict = scaler.inverse_transform(train_predict)
    y_train = scaler.inverse_transform([y_train])
    test_predict = scaler.inverse_transform(test_predict)
    y_test = scaler.inverse_transform([y_test])

    return y_test[0], test_predict[:,0]

# Example usage for solar power
y_solar_actual, y_solar_predicted = lstm_forecast(merged_data['Solar_MW'])
solar_lstm_rmse = rmse_score(y_solar_actual, y_solar_predicted)
print(f"Solar LSTM RMSE: {solar_lstm_rmse:.2f}%")


# Example usage for wind power (replace 'Solar_MW' with 'Wind_MW')
y_wind_actual, y_wind_predicted = lstm_forecast(merged_data['Wind_MW'])
wind_lstm_rmse = rmse_score(y_wind_actual, y_wind_predicted)
print(f"Wind LSTM RMSE: {wind_lstm_rmse:.2f}%")

# Example usage for demand (replace 'Solar_MW' with 'Demand_MW')
y_demand_actual, y_demand_predicted = lstm_forecast(merged_data['Demand_MW'])
demand_lstm_rmse = rmse_score(y_demand_actual, y_demand_predicted)
print(f"Demand LSTM RMSE: {demand_lstm_rmse:.2f}%")
````

````python
from pyomo.environ import *

# Define model
model = ConcreteModel()

# Given data
avg_solar_production = np.mean(y_solar_pred)
avg_wind_production = np.mean(y_wind_pred)
avg_daily_demand = np.mean(y_pred_demand)

grid_emission_factor = 0.5  # Emissions per kWh
solar_cost_per_kwh = 0.05
wind_cost_per_kwh = 0.07
natural_gas_cost_per_kwh = 0.15
grid_emission_penalty = 0.1  # Cost per emission unit

# Define variables
model.grid_supply = Var(bounds=(0, None), domain=NonNegativeReals)
model.solar_supply = Var(bounds=(0, avg_solar_production), domain=NonNegativeReals)
model.wind_supply = Var(bounds=(0, avg_wind_production), domain=NonNegativeReals)

# Objective function: Minimize total cost (including emissions)
def total_cost(model):
    return (
        model.grid_supply * natural_gas_cost_per_kwh +
        model.solar_supply * solar_cost_per_kwh +
        model.wind_supply * wind_cost_per_kwh +
        model.grid_supply * grid_emission_factor * grid_emission_penalty
    )
model.objective = Objective(rule=total_cost, sense=minimize)

# Constraint: Ensure total energy supply meets demand
model.demand_constraint = Constraint(expr=model.grid_supply + model.solar_supply + model.wind_supply >= avg_daily_demand)

# Solve the problem
solver = SolverFactory('cbc')  # Use GLPK solver (or 'gurobi' if available)
solver.solve(model)

# Output results
print("Optimal Solution Found")
print("Grid Supply:", value(model.grid_supply))
print("Solar Supply:", value(model.solar_supply))
print("Wind Supply:", value(model.wind_supply))
print("Total Cost:", value(model.objective))
````

````python
from scipy.optimize import linprog

# Given data
avg_solar_production = np.mean(y_solar_pred)
avg_wind_production = np.mean(y_wind_pred)
avg_daily_demand = np.mean(y_pred_demand)

solar_cost_per_kwh = 0.05
wind_cost_per_kwh = 0.07
natural_gas_cost_per_kwh = 0.15

# Objective function coefficients (costs)
c = [natural_gas_cost_per_kwh, solar_cost_per_kwh, wind_cost_per_kwh]

# Inequality constraints (A_ub * x <= b_ub)
A_ub = [
    [-1, -1, -1],  # -grid - solar - wind <= -avg_daily_demand
    [0, 1, 0],     # solar <= avg_solar_production
    [0, 0, 1]      # wind <= avg_wind_production
]
b_ub = [-avg_daily_demand, avg_solar_production, avg_wind_production]

# Bounds for each variable (non-negative)
x_bounds = [(0, None), (0, avg_solar_production), (0, avg_wind_production)]

# Solve the linear programming problem
result = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=x_bounds, method='highs')

# Check if the optimization was successful
if result.success:
    print("Optimal Solution Found")
    print(f"Grid Supply: {result.x[0]:.2f} kWh")
    print(f"Solar Supply: {result.x[1]:.2f} kWh")
    print(f"Wind Supply: {result.x[2]:.2f} kWh")
    print(f"Total Cost: ${result.fun:.2f}")
else:
    print("No optimal solution found")
````
